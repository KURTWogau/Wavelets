```python
data = pd.read_csv("2500yrData3.csv")
```

```python
x1 = data['Sr/Fe'].values
y1 = data['Ti'].values
t  = data['Yr BP'].values
```

```python
t
```

```python
x = wv.Wavepal(t,x1,"Age","Sr/Fe",t_units="yr",mydata_units="CPS")
```

```python
help(x.__init__)
```

```python
x.check_data()
```

```python
plot_timestep=x.plot_timestep()
#plot_timestep.show()
```

```python
plot_trend=x.plot_trend(pol_degree=-1)    # no trend
plot_trend.show()
plot_trend=x.plot_trend(pol_degree=0)    # constant trend
plot_trend.show()
plot_trend=x.plot_trend(pol_degree=5)    
plot_trend.show()
plot_trend=x.plot_trend(pol_degree=7)   
plot_trend.show()
plot_trend=x.plot_trend(pol_degree=9)    
plot_trend.show()
```
